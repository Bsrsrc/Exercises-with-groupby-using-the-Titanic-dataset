```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

---
#### Check the columns and missing values.
---

```python
df.columns
df.isnull().sum()
```

---
#### Fill the missing values in the 'age' variable with the median of the age variable.
---

```python
df["age"] = df["age"].fillna(df["age"].median())
df.head()
```

---
#### Is there a relationship between survival rates and passengers' ages in the Titanic dataset?
---

```python
df["age_cat"] = pd.cut(df["age"], bins=[1, 18, 30, 50, 100], labels=["0-18", "19-30", "31-50", "51+"])
df.head()
a = df.groupby("age_cat", observed=True).agg({"survived" : "sum"})
print(a)
```
